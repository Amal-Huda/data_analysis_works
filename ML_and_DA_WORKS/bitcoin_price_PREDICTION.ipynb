```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
bitdf=pd.read_csv("BitcoinPrice (4).csv")
```

```python
bitdf.head()
```

```python
bitdf.tail()
```

```python
bitdf.info()
```

```python
bitdf.drop(['Date'],1,inplace=True)
```

```python
bitdf.head()
```

```python
pt_days=30
bitdf['Predicted value']=bitdf[['Price']].shift(-pt_days)
```

```python
bitdf.head()
```

```python
bitdf.tail()
```

```python
x=np.array(bitdf.drop(['Predicted value'],1))
x=x[:len(bitdf)-pt_days]
           
```

```python
x.shape
```

```python
x
```

```python
y=np.array(bitdf['Predicted value'])
y=y[:-pt_days]
```

```python
y.shape
```

```python
y
```

```python
#split the data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train
```

```python
p_days_arr=np.array(bitdf.drop(['Predicted value'],1))[-pt_days:]
```

```python
print(p_days_arr)
```

```python
from sklearn.ensemble import RandomForestRegressor
Rf=RandomForestRegressor(n_estimators=1000,random_state=1)
Rf.fit(x_train,y_train)
```

```python
print("Random Forest accuracy:{:.2f}%".format(Rf.score(x_test,y_test)*100))
```

```python
rf_predict=Rf.predict(x_test)
```

```python
print(rf_predict)
```

```python
rf_pred_30=Rf.predict(p_days_arr)
```

```python
print(rf_pred_30)
```
